```python
from scipy import sparse
import pickle
import argparse
import logging
import pandas as pd
import numpy as np
import anndata
from scipy import sparse

from robustlink import utils
from robustlink import enhancer_gene_utils
```

# rna

```python
f = './data/rna_hvfeatures.cell'
cells = pd.read_csv(f, header=None)[0].values
f = './data/rna_hvfeatures.gene'
genes = pd.read_csv(f, header=None)[0].values
f = './data/rna_hvfeatures.npz'
mat = sparse.load_npz(f)
f = './data/rna_metadata.tsv'
meta = pd.read_csv(f, sep='\t')
assert np.all(meta['cell'].values == cells)
```

```python
adata = anndata.AnnData(
    X=mat,
    var=meta.set_index('cell'),
    obs=pd.DataFrame(index=genes),
)
```

```python
adata.write('./newdata/rna_hvfeatures.h5ad')
```
